## Cross-Validation

Statistical way to split in testing/training, used when evaluating model.

Simplest way to do it is through the $k$-Fold method, using sklearn's built-in methods.

```python
from sklearn.model_selection import cross_val_score, KFold
```

This is a great way to determine if any of the models are *overfitted*. 

It can be better to use shuffle with stratification, getting a better idea of how the model may perform across the entire dataset.

```python
from sklearn.model_selection import ShuffleSplit
```

With this, data is essentially being pulled in at random. This is the better way to split the data to get an idea of a model's accuracy.

## SVMs (Support Vector Machines) as Classifiers

Attempts to "chop-up" a $n$ dimensional feature plane by adding a hyper-parameter plane. Using linear algebra, can map feature dimensionality plane to map features into a larger dimensionality plane (can map to higher-dimensionality space) to determine a good hyperplane.

In other words, makes a linear hyperplane, mapping to higher dimensionality if hyperplane is insufficient.

Support Vectors are the points that lie closest to the hyperplane, aiding in linearly seperating them up.

## Neurel Nets (Regression and Classification)

### Typical NN (Multilayer Perceptron)

Takes in input layer and hidden layers. A complete graph from layer to layer. Not always true, but can be easiest to think of as one output layer (multiple inputs, one output). Each neuron/perceptron has a "tiny" learning function (starts at random values/random positions in learning function), and adjusts each perception to get closer to "Target" (loss) by throwing a "tensor" back through and adjusting all perceptrons.

### Activation Functions (Learning Functions)

Lives within each perceptron. Three major ones currently:

- Logistic Function (type of exponential function with two asymptopes, gentle learning curve)
- Relu (like rectifier function)
- Hyperbolic Tangent (Tanh, steeper learning curve)

Each have their own benefits.

### Usage

We used sklearn (interesting), can use both classifiers and regressors. 

Something cool we can do is plot y-predicted vs. y-test results to understand how things are being predicted/inaccurate. This can essentially tell us what the model is thinking and how it is being wrong.

Neurel Nets are very flexible, but very fragile. Therefore, it is very beneficial to run hyperparameter tuning (such as grid search) with them.
